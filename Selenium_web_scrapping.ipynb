```python
# Final step

from tqdm import tqdm
alldetails=[]
for i in tqdm(listoflinks):
    webD.get(i)
    nameofproduct=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[2]').text
    descriptionofprouct=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/p').text
    reviewofproduct=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[3]/p').text
    priceofproduct=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[1]').text
    tempJ={'nameofproduct':nameofproduct,
       'priceofproduct':priceofproduct,
       'reviewofproduct':reviewofproduct,
       'descriptionofprouct':descriptionofprouct,
       'linkofproduct':i}
    alldetails.append(tempJ)
```
